```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
kategorik_sutunlar = ["gender", "part_time_job", "extracurricular_participation"]

for sutun in kategorik_sutunlar:
    df[sutun] = pd.factorize(df[sutun])[0]  ## factorize"yi ilk defa kullandim burada yapay zekadan yardim alarak boyle bir attiribute oldugunu buldum
```

```python
ordinal_esleme = {
    "diet_quality": {"Poor": 0, "Average": 1, "Good": 2},
    "parental_education_level": {"High School": 0, "Bachelor": 1, "Master": 2, "PhD": 3},
    "internet_quality": {"Low": 0, "Medium": 1, "High": 2}
}

for sutun, esleme in ordinal_esleme.items():
    df[sutun] = df[sutun].map(esleme)

df.drop("internet_quality", axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```
